## 2.2 Outlier detection

```python
def detect_outliers(df, n, features):
    """
    Takes a dataframe df of features and returns a list of the indices
    corresponding to the observations containing morethan n outliers according to the Tukey method.
    """
    outlier_indices = []

    # iterate over features(columns)
    for col in features:
        # 1st quartile (25%)
        Q1 = np.percentile(df[col], 25)
        # 3rd quartile (75%)
        Q3 = np.percentile(df[col], 75)
        # Interquartile range (IQR)
        IQR = Q3 - Q1

        # outlier step
        outlier_step = 1.5 * IQR

        # Determine a list of indices of outliers for feature col
        outlier_list_col = df[
            (df[col] < Q1 - outlier_step) | (df[col] > Q3 + outlier_step)
        ].index

        # append the found outlier indices for col to the list of outlier indices
        outlier_indices.extend(outlier_list_col)

    # select observations containing more than n outliers
    outlier_indices = Counter(outlier_indices)
    multiple_outliers = list(k for k, v in outlier_indices.items() if v > n)

    return multiple_outliers


# detect outliers from Age, SibSp, Parch and Fare
Outliers_to_drop = detect_outliers(train, 2, ["Age", "SibSp", "Parch", "Fare"])
Outliers_to_drop
```

Since outliers can have a dramatic effect on the prediction(especially for regression problems), i choosed to manage them.

I used the Tukey method(Tukey JW., 1977) to detect outliers which defines an interquartile range comprised between the 1st and 3rd quartile of the distribution values(IQR). An outliers is a row that have a feature value outside the (IQR +- an outlier step).

I decided to detect outliers from the numerical values features(Age, SibSp, Parch and Fare). Then, I considered outliers as rows that have at least two outlied numerical values.

#### Tukey Method

**Tukey Method**는 John Tukey가 제안한 이상치(outlier) 탐지 방법으로, **사분위수**(quartiles)를 이용한 간단한 이상치 탐지 기법입니다.

##### 핵심 개념
- **Q1 (제 1 사분위수)**: 하위 25%에 해당하는 값
- **Q3 (제 3 사분위수)**: 상위 25%에 해당하는 값
- **IQR (Interquartile Range)**: 중간 50%의 범위, \( IQR = Q3 - Q1 \)

##### 이상치 탐지 기준
- **일반적인 이상치**: 
  - 하위 경계: \( Q1 - 1.5 \times IQR \)
  - 상위 경계: \( Q3 + 1.5 \times IQR \)
- **극단적인 이상치**:
  - 하위 경계: \( Q1 - 3 \times IQR \)
  - 상위 경계: \( Q3 + 3 \times IQR \)

##### 절차
1. 데이터 정렬
2. **Q1**, **Q3** 계산 및 **IQR** 산출
3. 경계를 계산하여 벗어난 값은 이상치로 간주

##### 장점
- **간단하고 직관적**: 계산이 쉽고 시각적으로 이해하기 용이
- **통계적 가정이 없음**: 데이터 분포에 대한 가정이 없어 다양한 데이터셋에 적용 가능

##### 단점
- **극단적인 분포에 취약**: 비대칭적이거나 극단값이 많은 경우 한계가 있음

### Fare

```python
dataset["Fare"].isnull().sum()
```

```python
# Fill Fare missing values with the median value
dataset["Fare"] = dataset["Fare"].fillna(dataset["Fare"].median())
```

Since we have one missing value, I decided to fill it with the median value which will not have an important effect on the prediction.

```python
# Explore Fare distribution
g = sns.histplot(
    dataset["Fare"],
    kde=True,
    color="m",
    label="Skewness : %.2f" % (dataset["Fare"].skew()),
)
g = g.legend(loc="best")
```

As we can see, Fare distribution is very skewed. This can lead to overweight very high values in the model, even if it is scaled.

In this case, it is better to transform it with the log function to reduce this skew.

Skewness가 큰 데이터는 모델에서 극단값(outliers)이 모델의 가중치 계산에 과도하게 영향을 줄 수 있습니다. 위 Fare의 경우에 Fare는 대부분 낮은 값들로 구성되지만 일부 높은 값들이 있어 이 값들을 과도하게 중요하게 여기게 될 수 있습니다. 심지어 데이터를 스케일링(정규화)해도 이런 극단값이 영향을 미칠 수 있습니다.

로그 변환은 왜도가 큰 데이터 분포를 더 대칭적인 분포로 변환하는 데 자주 사용됩니다. 높은 값은 작게 변환되고, 작은 값은 비교적 덜 변환되므로 극단값의 영향이 줄어들고, 데이터가 좀 더 균등하게 분포하게 됩니다.

```python
# Apply log to Fare to reduce skewness distribution
dataset["Fare"] = dataset["Fare"].map(lambda i: np.log(i) if i > 0 else 0)
```

```python
g = sns.histplot(
    data=dataset["Fare"],
    kde=True,
    color="b",
    label="Skewness : %.2f" % (dataset["Fare"].skew()),
)
g = g.legend(loc="best")
```

Skewness is clearly reduced after the log transformation.

# 4. Filling missing Values

## 4.1 Age

As we see, Age column contains 256 missing values in the whole dataset.

Since there is subpopulations that have more chance to survive(children for example), it is preferable to keep the age feature and to impute the missing values.

To address, this problem, I looked at the most correlated features with Age(Sex, Parch, Pclass and SibSP).

```python
# Explore Age vs Sex, Parch, Pclass and SibSp
g = sns.catplot(data=dataset, x="Sex", y="Age", kind="box", hue="Sex")
g = sns.catplot(data=dataset, x="Sex", y="Age", kind="box", hue="Pclass")
g = sns.catplot(data=dataset, x="Parch", y="Age", kind="box", hue="Parch")
g = sns.catplot(data=dataset, x="SibSp", y="Age", kind="box", hue="SibSp")
```

Age distribution seems to be the same in Male and Femaile subpopulations, so Sex is not informative to predict Age.

However, 1st class passengers are older than 2nd class passengers who are also older than 3rd class passengers.

Moreover, the more a passenger has parents/children the older he is and the more a passenger has siblings/spouses the younger he is.

```python
# convert Sex into categorical value 0 for male and 1 for female
dataset["Sex"] = dataset["Sex"].map({"male": 0, "female": 1})
```

```python
g = sns.heatmap(
    dataset[["Age", "Sex", "SibSp", "Parch", "Pclass"]].corr(), cmap="BrBG", annot=True
)
```

The correlation map confirms that catplots observations except for Parch. Age is not correlated with Sex, but is negatively correlated with Pclass, Parch and SibSp.

In the plot of Age in function of Parch, Age is growing with the number of parents/children. But the general correlation is negative.

So I decided to use SibSp, Parch and Pclass in order to impute the missing ages.

The strategy is to fill Age with the median age of similar rows according to Pclass, Parch and SibSp.

```python
# Filling missing value of Age

# Fill Age with the median age of similar rows according to Pclass, Parch and SibSp
# Index of NaN age rows
index_NaN_age = list(dataset["Age"][dataset["Age"].isnull()].index)

for i in index_NaN_age:
    age_med = dataset["Age"].median()
    age_pred = dataset["Age"][
        (
            (dataset["SibSp"] == dataset.iloc[i]["SibSp"])
            & (dataset["Parch"] == dataset.iloc[i]["Parch"])
            & (dataset["Pclass"] == dataset.iloc[i]["Pclass"])
        )
    ].median()

    if not np.isnan(age_pred):
        dataset.loc[i, "Age"] = age_pred
    else:
        dataset.loc[i, "Age"] = age_med
```

```python
g = sns.catplot(data=train, x="Survived", y="Age", kind="box", hue="Survived")
g = sns.catplot(data=train, x="Survived", y="Age", kind="violin", hue="Survived")
```

No difference between median value of age in survived and not survived subpopulation.

But in the violin plot of survived passengers, we still notice that very young passengers have higher survival rate.

```python
# Create a family size descriptor from SibSp and Parch
dataset["Fsize"] = dataset["SibSp"] + dataset["Parch"] + 1
```

#### RandomForestClassifier vs ExtraTreesClassifier

##### 공통점
- 둘 다 다수의 **결정 트리**를 사용하는 **앙상블 기법**.
- 노드 분할 시 무작위로 선택된 일부 특성들만 고려하여 **트리 간 다양성**을 높임.
- 각 트리의 결과를 결합해 최종 예측을 도출.

##### 차이점

| 특성                          | **RandomForestClassifier**                                     | **ExtraTreesClassifier**                                     |
| ----------------------------- | --------------------------------------------------------------- | ------------------------------------------------------------ |
| **데이터 샘플링**              | 각 트리가 **부트스트랩 샘플링**된 데이터를 사용해 학습           | **전체 훈련 데이터셋**을 사용해 학습                         |
| **노드 분할 기준**            | 무작위로 선택된 일부 특성에서 **최적의 분할**을 찾음            | 무작위로 선택된 특성에서 **무작위 분할 포인트**를 설정       |
| **학습 속도**                  | 상대적으로 **느림** (최적의 분할을 계산함)                      | 상대적으로 **빠름** (분할 기준을 무작위로 선택)              |
| **무작위성의 정도**            | **적음** (특성만 무작위, 분할은 최적화)                        | **많음** (특성과 분할 기준 모두 무작위)                      |
| **성능**                       | 일반적으로 더 **높은 성능**을 보일 수 있음                      | 더 빠른 학습 속도를 제공하나, 성능은 특정 데이터에서 다소 **떨어질 수 있음** |

##### 요약
- **RandomForestClassifier**: 데이터 샘플링과 특성 선택에서 무작위성을 도입하며, **최적의 분할**을 계산해 학습. 성능이 좋고, 다소 느릴 수 있음.
- **ExtraTreesClassifier**: 특성과 분할 기준 모두 무작위로 설정하여 학습. 더 빠르지만 성능이 떨어질 수 있음.


#### AdaBoostClassifier vs GradientBoostingClassifier

##### 공통점
- 둘 다 **부스팅(Boosting)** 기법을 사용하는 **앙상블 학습** 방법.
- 여러 **약한 학습자(Weak Learner)**를 순차적으로 학습하여 성능을 향상시킴.
- 주로 **결정 트리(Decision Tree)**를 약한 학습자로 사용.

---

##### AdaBoostClassifier

###### 학습 과정
1. **초기 가중치 부여**: 모든 데이터 포인트에 동일한 가중치를 부여.
2. **첫 번째 학습자 학습**: 약한 학습자가 데이터를 학습하고 예측 수행.
3. **예측 평가 및 가중치 업데이트**: 잘못 예측된 데이터에 더 큰 가중치를 부여.
4. **다음 학습자 학습**: 두 번째 학습자는 가중치가 업데이트된 데이터를 사용해 학습.
5. **반복**: 이 과정을 설정된 학습자 수만큼 반복.
6. **최종 예측**: 학습자들의 예측을 **가중치 기반으로 결합**해 최종 예측 도출.

###### 주요 특징
- 각 단계에서 **잘못 예측된 데이터**에 더 많은 가중치를 부여하여 **오류를 보정**.
- 최종 예측은 **가중치 기반 투표** 방식으로 결정.
- 간단하고 빠른 학습, 하지만 **노이즈**에 민감할 수 있음.

---

##### GradientBoostingClassifier

###### 학습 과정
1. **초기 모델 학습**: 첫 번째 모델이 데이터를 학습하여 예측.
2. **잔차 계산**: 예측과 실제 값 간의 차이(잔차)를 계산.
3. **잔차 기반 학습**: 두 번째 학습자는 **잔차(오류)를 줄이도록** 학습.
4. **모델 업데이트**: 새로운 학습자가 기존 모델을 보완.
5. **반복**: 설정된 학습자 수만큼 반복하며, 각 학습자는 잔차를 줄이도록 학습.
6. **최종 예측**: 모든 학습자의 결과를 결합하여 최종 예측 도출.

###### 주요 특징
- 각 단계에서 **잔차(오류)를 줄이는 방식**으로 학습.
- **손실 함수**를 최적화하여 성능을 점진적으로 향상.
- **더 복잡한 학습** 과정과 **긴 학습 시간**, 하지만 **높은 성능** 발휘 가능.

---

##### 차이점 비교

| 특성                        | **AdaBoostClassifier**                                       | **GradientBoostingClassifier**                             |
| --------------------------- | ------------------------------------------------------------ | ---------------------------------------------------------- |
| **학습 방식**                | 이전 모델의 **잘못된 데이터**에 가중치 부여                  | 이전 모델의 **잔차(오류)를 줄이는 방식**                   |
| **결합 방식**                | **가중치 기반 투표**                                         | **잔차 기반 업데이트**                                     |
| **손실 함수 최적화**         | 사용하지 않음                                                | **손실 함수의 기울기**를 최소화하며 학습                   |
| **약한 학습자**              | 주로 **깊이가 얕은 결정 트리** (Stump)                       | **결정 트리** (더 깊은 트리 사용 가능)                     |
| **학습 속도**                | 비교적 빠름                                                 | 비교적 느림                                                |
| **성능**                     | 단순하고 빠르게 사용 가능하지만, 데이터 노이즈에 민감할 수 있음 | 더 복잡한 데이터에 적합하며, **높은 성능**을 발휘함        |

---

##### 요약
- **AdaBoostClassifier**는 이전 학습자의 **잘못된 예측에 가중치를 부여**하여 오류를 보정하고, 각 학습자의 결과를 **가중치 기반 투표**로 결합.
- **GradientBoostingClassifier**는 이전 모델의 **잔차를 줄이기 위해 새로운 모델을 학습**하며, **손실 함수를 최적화**하여 성능을 점진적으로 개선.
- AdaBoost는 빠르고 간단하지만 노이즈에 민감할 수 있고, Gradient Boosting은 더 복잡하고 느리지만 성능이 뛰어남.

### 6.1.2 Hyperparameter tuning for best models
I performed a grid search optimization for AdaBoost, ExtraTrees, RandomForest, GradientBoosting and SVC classifiers.

This step can take a long time.

```python
# META MODELING WITH ADABOOST, RF, EXTRATREES and GRADIENTBOOSTING

# Adaboost
DTC = DecisionTreeClassifier()

adaDTC = AdaBoostClassifier(DTC, random_state=7)

ada_param_grid = {
    "estimator__criterion": ["gini", "entropy"],  # 결정 트리 분류기 분할 기준
    "estimator__splitter": ["best", "random"],  # 데이터 분할 시 노드 분할 방식
    "algorithm": ["SAMME"],  # AdaBoost에서 사용할 알고리즘
    "n_estimators": [1, 2],  # 부스팅을 위해 생성할 약한 학습자(결정 트리) 수
    "learning_rate": [0.0001, 0.001, 0.01, 0.1, 0.2, 0.3, 1.5],  # 학습률
}

gsadaDTC = GridSearchCV(
    adaDTC, param_grid=ada_param_grid, cv=kfold, scoring="accuracy", n_jobs=4, verbose=1
)

gsadaDTC.fit(X_train, Y_train)

ada_best = gsadaDTC.best_estimator_
```

```python
gsadaDTC.best_score_
```

```python
# ExtraTrees
ExtC = ExtraTreesClassifier()

# Search grid for optimal parameters
ex_param_grid = {
    "max_depth": [None],  # 최대한 깊이 성장할 수 있음
    "max_features": [1, 3, 10],  # 각 노드에서 분할 시도 시 고려할 최대 특징 수
    "min_samples_split": [2, 3, 10],  # 노드 분할 시 필요한 최소 샘플 수
    "bootstrap": [False],  # 부트스트랩 샘플링 사용 여부
    "n_estimators": [100, 300],  # 생성할 트리 수
    "criterion": ["gini"],  # 분할 품질 평가 기준
}

gsExtC = GridSearchCV(
    ExtC, param_grid=ex_param_grid, cv=kfold, scoring="accuracy", n_jobs=4, verbose=1
)

gsExtC.fit(X_train, Y_train)

ExtC_best = gsExtC.best_estimator_

gsExtC.best_score_
```

```python
# RFC Parameters tunning
RFC = RandomForestClassifier()

# Search grid for optimal params
rf_param_grid = {
    "max_depth": [None],  # 최대한 깊이 성장할 수 있음
    "max_features": [1, 3, 10],  # 각 노드 분할 시도 시 고려할 최대 특징 수
    "min_samples_split": [2, 3, 10],  # 노드 분할 시 필요한 최소 샘플 수
    "min_samples_leaf": [1, 3, 10],  # leaf node에 있어야 하는 최소 샘플 수
    "bootstrap": [False],  # 부트스트랩 샘플링 사용 여부
    "n_estimators": [100, 300],  # 생성할 트리 수
    "criterion": ["gini"],  # 분할 품질 평가 기준
}

gsRFC = GridSearchCV(
    RFC, param_grid=rf_param_grid, cv=kfold, scoring="accuracy", n_jobs=4, verbose=1
)

gsRFC.fit(X_train, Y_train)

RFC_best = gsRFC.best_estimator_

gsRFC.best_score_
```

```python
# Gradient boosting tunning

GBC = GradientBoostingClassifier()
gb_param_grid = {
    "loss": ["log_loss"],  # 손실 함수
    "n_estimators": [100, 200, 300],  # 생성할 약한 학습자(결정 트리) 수
    "learning_rate": [0.1, 0.05, 0.01],  # 학습률
    "max_depth": [4, 8],  # 각 결정 트리의 최대 깊이
    "min_samples_leaf": [100, 150],  # leaf node의 최소 샘플 수
    "max_features": [0.3, 0.1],  # 각 분할 시 사용할 최대 특징 수 비율
}

gsGBC = GridSearchCV(
    GBC, param_grid=gb_param_grid, cv=kfold, scoring="accuracy", n_jobs=4, verbose=1
)

gsGBC.fit(X_train, Y_train)

GBC_best = gsGBC.best_estimator_

gsGBC.best_score_
```

```python
# SVC classifier
SVMC = SVC(probability=True)  # 예측 결과를 확률로 출력할 수 있도록 설정
svc_param_grid = {
    "kernel": ["rbf"],  # 사용할 커널 함수
    # rbf 커널의 파라미터로, 데이터 포인트의 영향력 결정. 값이 작을수록 멀리 있는 데이터포인트들의 영향이 커짐
    "gamma": [
        0.001,
        0.01,
        0.1,
        1,
    ],
    # 오차 허용 범위. 값이 클수록 오차를 허용하지 않으려 함
    "C": [1, 10, 50, 100, 200, 300, 1000],
}

gsSVMC = GridSearchCV(
    SVMC, param_grid=svc_param_grid, cv=kfold, scoring="accuracy", n_jobs=4, verbose=1
)

gsSVMC.fit(X_train, Y_train)

SVMC_best = gsSVMC.best_estimator_

gsSVMC.best_score_
```

### 6.1.3 Plot learning curves

Learning curves are a good way to see the overfitting effect on the training set and the effect of the training size on the accuracy.

```python
def plot_learning_curve(
    estimator,
    title,
    X,
    y,
    ylim=None,
    cv=None,
    n_jobs=-1,
    train_sizes=np.linspace(0.1, 1.0, 5),
):
    """Generate a simple plot of the test and training learing curve"""
    plt.figure()
    plt.title(title)
    if ylim is not None:
        plt.ylim(*ylim)
    plt.xlabel("Training examples")
    plt.ylabel("Score")
    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, y, cv=cv, n_jobs=n_jobs, train_sizes=train_sizes
    )
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    plt.grid()

    plt.fill_between(
        train_sizes,
        train_scores_mean - train_scores_std,
        train_scores_mean + train_scores_std,
        alpha=0.1,
        color="r",
    )
    plt.fill_between(
        train_sizes,
        test_scores_mean - test_scores_std,
        test_scores_mean + test_scores_std,
        alpha=0.1,
        color="g",
    )
    plt.plot(train_sizes, train_scores_mean, "o-", color="r", label="Training score")
    plt.plot(
        train_sizes, test_scores_mean, "o-", color="g", label="Cross-validation score"
    )
    plt.legend(loc="best")
    return plt


g = plot_learning_curve(
    gsRFC.best_estimator_, "RF learning curves", X_train, Y_train, cv=kfold
)
g = plot_learning_curve(
    gsExtC.best_estimator_, "ExtraTrees learning curves", X_train, Y_train, cv=kfold
)
g = plot_learning_curve(
    gsSVMC.best_estimator_, "SVC learning curves", X_train, Y_train, cv=kfold
)
g = plot_learning_curve(
    gsadaDTC.best_estimator_, "AdaBoost learning curves", X_train, Y_train, cv=kfold
)
g = plot_learning_curve(
    gsGBC.best_estimator_,
    "GradientBoosting learning curves",
    X_train,
    Y_train,
    cv=kfold,
)
```

- **RandomForest**: 약간의 과적합 가능성이 있지만, 비교적 안정된 성능
- **ExtraTrees**: 훈련 점수와 검증 점수 차이가 커서 과적합 가능성 높음
- **SVC**: 처음엔 과적합 있었으나, 데이터 늘어남에 따라 점차 안정화
- **AdaBoost**: 심각한 과적합
- **GradientBoosting**: 훈련 및 검증 점수가 거의 동일, 과적합이나 과소적합 없이 좋은 성능

### 6.1.4 Feature importance of tree based classifiers
In order to see the most informative features for the prediction of passengers survival, I displayed the feature importance for the 4 tree based classifiers.

```python
type(ada_best).__name__
```

```python
nrows = ncols = 2
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, sharex="all", figsize=(15, 15))

classifiers = [ada_best, ExtC_best, RFC_best, GBC_best]

nclassifier = 0
for row in range(nrows):
    for col in range(ncols):
        classifier_index = row * ncols + col
        name = type(classifiers[classifier_index]).__name__
        classifier = classifiers[classifier_index]
        # feature_importances_를 내림차순 정렬 후 상위 40개 인덱스 추출
        indices = np.argsort(classifier.feature_importances_)[::-1][:40]
        g = sns.barplot(
            y=X_train.columns[indices][:40],
            x=classifier.feature_importances_[indices][:40],
            orient="h",
            ax=axes[row][col],
        )
        g.set_xlabel("Relative importance", fontsize=12)
        g.set_ylabel("Features", fontsize=12)
        g.tick_params(labelsize=9)
        g.set_title(name + " feature importance")
        nclassifier += 1
```

I plot the feature importance for the 4 tree based classifiers(Adaboost, ExtraTrees, RnadomForest and GradientBoosting).

We note that the four classifiers have different top features according to the relative importance. It means that their predictions are not based on the same features. Nevertheless, they share some common important features for the classification, for example "Fare", "Title_2", "Age" and "Sex".

Title_2 which indicates the Mrs/Mlle/Mme/Miss/Ms category is highly correlated with Sex.

We can say that:
- Pc_1, Pc_2, Pc_3 and Fare refer to the general social standing of passengers.
- Sex and TItle_2(Mrs/Mlle/Mme/Miss/Ms) and Title_3(Mr) refer to the gender.
- Age and Title_1(Master) refer to the age passengers.
- Fsize, LargeF, MedF, Single refer to the size of the passenger family.

**According to the feature importance of this 4 classifiers, the prediction of the survival seems to be more associated with the Age, the Sex, the family size and the social standing of the passengers more than the location in the boat.**

## 6.2 Ensemble modeling

### 6.2.1 Combining models

I choosed a voting classifier to combine the predictions coming from the 5 classifiers.

I preferred to pass the argument "soft" to the voting parameter to take into account the probability of each vote.

hard voting은 각 모델의 예측 클래스 레이블을 기반으로 다수결 투표를 하는 것이고, soft voting은 각 모델의 예측 확률을 기반으로 평균을 계산해 최종 예측을 결정한다.

```python
votingC = VotingClassifier(
    estimators=[
        ("rfc", RFC_best),
        ("extc", ExtC_best),
        ("svc", SVMC_best),
        ("adac", ada_best),
        ("gbc", GBC_best),
    ],
    voting="soft",
    n_jobs=4,
)

votingC = votingC.fit(X_train, Y_train)
```
