```python
from PIL import Image
import torch
import torchvision
import torchvision.transforms as T

# 모델 로드
weights = torchvision.models.detection.FasterRCNN_ResNet50_FPN_Weights.DEFAULT
model = torchvision.models.detection.fasterrcnn_resnet50_fpn(weights=weights)
model.eval()
```

```python
# 이미지 전처리 함수
def preprocess_image(image):
    transform = T.Compose([T.ToTensor()])
    return transform(image)
```

```python
# 예시 이미지 경로
image_path = "./1.jpg"
image = Image.open(image_path)
image_tensor = preprocess_image(image)
```

```python
with torch.no_grad():
    predictions = model([image_tensor])
```

```python
import matplotlib.pyplot as plt
from PIL import ImageDraw, ImageFont

def plot_image_with_boxes(image_pil, predictions, threshold=0.5):
    # 예측 결과에서 바운딩 박스와 라벨 추출
    boxes = predictions[0]['boxes']
    labels = predictions[0]['labels']
    scores = predictions[0]['scores']

    # 이미지 그리기 객체 생성
    draw = ImageDraw.Draw(image_pil)

    # 박스 그리기
    for box, label, score in zip(boxes, labels, scores):
        if score >= threshold:  # 임계값 이상의 확신도를 가진 경우만 표시
            box = box.detach().cpu().numpy()

            # 박스 그리기 (좌상단(x1, y1) -> 우하단(x2, y2))
            draw.rectangle([(box[0], box[1]), (box[2], box[3])], outline="green", width=2)
            
            # 라벨과 점수 그리기
            draw.text((box[0], box[1] - 10), f"Label: {label}, Score: {score:.2f}", fill="green")

    # 시각화
    plt.figure(figsize=(18, 12))
    plt.imshow(image_pil)
    plt.axis("off")
    plt.show()
```
